## Introduction

Mayflies are small aquatic insects (typically .6 to 2.8cm) which belong to a group of insects also containing dragonflies and damselflies. They have very short lifespans and they emerge in large numbers during the summer months. Mayflies usually inhabit clean, unpolluted environments and have quite a uniqe reproduction method in which the females deposit eggs in the bottom of a river and then the babies emerge after about a year. The babies emerge as adults, and they mate with each other just to repeat the cycle over and over again. Adults only live for about a day after they emerge, so most of their life is spent in the nymph stage (underwater). 

Mathematical models can be useful for predicting how populations will behave in the future so that we can look at short run and long term behavior. In this case, we are interested in modeling the population of mayflies in an arbitrary environment. This model may not be extremely accurate to what happens in the real world because we are not accounting for more complex factors, we are making many assumptions to begin with. 

Assuming the population of mayflies in a specific year is dependent on the size of the population in the previous year, $x_{n} = g(x_{n-1})\cdot x_{n-1}$ represents the size of the population in year $n$. $g(x_{n-1})$ represents the growth rate of the population and $x_{n-1}$ represents the population in the previous year. 

The carrying capacity is the maximum number of organisms an environment can support and $g(x)$ will be a decreasing function because as the population increases, it approaches the carrying capacity of the environment. Also, as the population of mayflies increases, there will be more competition for resources which is another reason why $g(x)$ is a decreasing function. 

We will use $M$ to represent the carrying capacity of the environment. A good way to represent the growth rate in terms of the population at a given year and $M$ would be the function $g(x) = a(M-x)$. Graphically, $a$ is a positive constant and this forms a line which decreases to the point $M$. If we replace our new $g(x)$ into our first equation, we end up with $x_n = a(M-x_{n-1})x_{n-1}$. Now we have two parameters $M$ and $a$ which we can combine into one variable $y_n$, namely $y_n = \dfrac{x_n}{M}$ so now our equation becomes $y_n = b(1-y_{n-1})y_{n-1}$.

In our function, $0 \leq y_n \leq 1$ and $0 \leq b \leq 4$.

Now, the goal for this project is to explore what different values for $y_0$, the initial value of percentage of the population, and $b$ produce. We will explore the short term behavior for the mayfly model as well as the long term behavior, and graph values of $b$ which we will later look at.

## Content

First, we will define a function `mayfly(b,y0)` which will return the population at the next day given the previous day using our mayfly equation $y_n=b(1-y_{n-1}) y_{n-1}$.

```python
def mayfly(b,y0): # defining the mayfly function
    yn=b*(1-y0)*y0 # b must be in [0,4] and y0 is the percent of the population at day 0
    return yn
```

```python
mayfly(2,.4) # testing the function to see if it works
```

```python
mayfly(3,.9) # testing the function to see if it works
```

## Graphing

Now that we have a function that successfully returns the next day's population given the previous day, we can graph this function and explore what changing the $b$ and $y_0$ results in. We will first explore the transient behavior, or what happens in the short run (let's say 20 days).

```python
import matplotlib.pyplot as plt # importing the plotting module and giving it a nickname
import numpy as np # importing the numpy module and giving it a nickname
```

```python
days=20 # 20 days
b=1 # b value
y0=.4 # initial value of the percentage of the population
mylist1=[y0]
for n in range(days): # for loop which loops through the days
    yn=mayfly(b,y0) # defining yn to be the value we attain from our mayfly function
    mylist1.append(yn) # append yn to mylist1
    y0=yn # our previous value of y will help us determine the expected value 
plt.title('b=1, y0=.4') # title
plt.xlabel('Days') # x axis label
plt.ylabel('Percent Population') # y axis label
plt.plot(mylist1,'b-*')
plt.show()
```

This is good, however we can make a much more efficient display by showing many graphs at once using subplots.

```python
days=20 # 20 days
blist=[1,2.5,3.75,4] # list of b values
ylist=[.1,.2,.7,.85] # list of y0 values

plt.figure(figsize=(15,15)) # adjusting the size of the graphs

k=1 # starting point
for b in blist: # for loop which loops through elements in blist
    for y0 in ylist: # for loop to loop through elements of ylist
        plt.subplot(4,4,k) # making a 4x4 subplot
        mylist=[y0] # new list to append our mayfly function values into
        for n in range(days): # looping through the days
            yn=mayfly(b,y0) # defining yn to be whatever our mayfly function gives
            mylist.append(yn) # add yn to mylist
            y0=yn # previous y value will help us determine the expected value
        plt.plot(mylist,'b-*')
        plt.subplots_adjust(wspace=.25,hspace=.35)
        plt.title(f'b={b}, y0={mylist[0]}') # title
        plt.xlabel('Days') # x axis label
        plt.ylabel('Percent Population') # y axis label
        k+=1 # iterates next step in the for loop
        
plt.show()
```

In general, altering the $b$ value has a more significant impact on what the graph looks like than altering the $y_0$ value.

## *b* Value Graph

Let's try to find where the boundary is for a b value that produces a graph which tends to a single point and a graph that will never tend to a single point. We will hold the $y_0$ value constant and go through different values of $b$.

```python
for b in np.linspace(0,4,600): # plotting b values from 0 to 4 with 600 steps
    bx=b*np.ones(100) # list of x coords for future plot
    by=[] # we will call an empty list by
    y0=.5 # initial value of percent of population
    for d in range(0,601): # loop through values 0-600
        ycoord=mayfly(b,y0) # set y coord = mayfly function value
        by.append(ycoord) # add y coord to by
        y0=ycoord
    byy=by[500:600] # a slice of an area of heavy concentration
    plt.xlabel('b value') # x axis label
    plt.ylabel('Percentage of Population') # y axis label
    plt.title('b Between 0 and 4') # title
    plt.plot(bx,byy,'b.',alpha=.5,ms=.3) # plots the graph, alpha is the transparency parameter 
plt.show()
```

We can immediately notice that there is an upward but decreasing trend from 1 to 3 and then it seems to turn into organized chaos.

## Conclusion

In most cases in the long run, we can infer that populations stabilize after a certain amount of time. You can change the initial population as much as you want, but in the long run the population of mayflies will reach an "equilibrium population" and stabilize itself. In our $b$ value graph, once the $b$ value reaches 3, the graph splits in 2, meaning that for $b$ values above 3 the population does not converge to a single point as they do with $b$ values less than 3. The graph oscillates back and forth between two values until a $b$ value of about 3.45, where the "fork" splits again. Now, periods are going through a 4 year cycle before repeating and this is known as a **period doubling bifurcation**. The rate of period doubling bifurcations occurs at an exponentially increasing rate as $b \to 4$. For the second, third, and fourth $b$ value graphs I have tried to start the range right at the next "fork" in the graph. In conclusion, we have successfully graphed what our mayfly model looks like in the short run and long run for different values of $b$ and $y_0$ and found out that altering the $b$ value is more significant. Because of this, we then graphed the $b$ values and found out some very interesting patterns about the model. It is quite interesting how the graph forks at an exponentially increasing rate until we hit our last value of $b$, 4.
